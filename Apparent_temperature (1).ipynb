# Data cleaning

#### checking the unique values 

```python
data.apply(lambda x: len(x.unique()))
```

## Replacing the incorrect values with appropriate values.

#### Humidity cannot be 74 so replacing with 0.74

```python
data['Humidity'].replace(74,0.74,inplace=True)
```

```python
data[['Humidity']].describe()
```

#### Temperature is 3014 which is inappropriate temperature cannot be 3014 degress so replacing with 13.14

```python
data['Temperature (C)'].replace(3014,13.14,inplace=True)
```

```python
data[['Temperature (C)']].describe()
```

#### Wind speed cannot be 1444 km/h as its speed can destroy atmospheric normal  going.so,replacing it with 14.44 km/h

```python
data['Wind Speed (km/h)'].replace(1444,14.44,inplace=True)
```

```python
data[['Wind Speed (km/h)']].describe()
```

#### summary value is 14 which is inapproprite so we delete row

```python
data.query('Summary==14')
```

As we observe that Loud Cover having entire column as zero.Visibility are not required as they does not have effect on apparent temperature droping this 2 columns.

```python
data = data.drop(['Loud Cover','Visibility (km)'], axis=1)
```

```python
data.shape
```

As we have droped 2 columns the number of columns has decreased from 12 to 10.

# Apparent Temperature vs Hours in Formatted date
This graph clearly shows the apparent temperature at a particular hour of all the days in the given dataset.

```python
from datetime import datetime
```

```python
data['Formatted Date']=pd.to_datetime(data['Formatted Date'],utc=True)
```

```python
data['Formatted Date']=data['Formatted Date'].astype(str).str[:-6]
```

```python
m=pd.to_datetime(data['Formatted Date'])
```

```python
data['month']=m.dt.month
data['year']=m.dt.year
data['hour']=m.dt.hour
data.groupby('hour')['Apparent Temperature (C)'].mean().to_frame().plot(kind='bar')
plt.title('Apparent Temperature vs Hours in Formatted date')
plt.show()
```
